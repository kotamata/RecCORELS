```python
import pandas as pd
import numpy as np
import re
import pickle
```

## データの前処理
- データをone-hot encoding（def convert）
- 

```python
train = pd.read_csv("compas_train.csv")
train.head(10)
```

```python
test = pd.read_csv("compas_test.csv")
test.head(10)
```

```python
def convert(df):
    df["sex:Male"] = (df.sex=="Male")
    df["sex:Female"] = (df.sex=="Female")
    
    df["age:18-20"] = (df.age=="18-20")
    df["age:21-22"] = (df.age=="21-21")
    df["age:23-25"] = (df.age=="23-25")
    df["age:26-45"] = (df.age=="26-45")
    df["age:>45"]   = (df.age==">45")
    
    df["juvenile-felonies:=0"] = (df["juvenile-felonies"]=="=0")
    df["juvenile-felonies:>0"] = (df["juvenile-felonies"]==">0")
    
    df["juvenile-misdemeanors:=0"] = (df["juvenile-misdemeanors"]=="=0")
    df["juvenile-misdemeanors:>0"] = (df["juvenile-misdemeanors"]==">0")
    
    df["juvenile-crimes:=0"] = (df["juvenile-crimes"]=="=0")
    df["juvenile-crimes:>0"] = (df["juvenile-crimes"]==">0")
    
    df["priors:=0"]  = (df["priors"]=="=0")
    df["priors:=1"]  = (df["priors"]=="=1")
    df["priors:2-3"] = (df["priors"]=="2-3")
    df["priors:>3"]  = (df["priors"]==">3")
    
    df["current-charge-degree:Felony"]      = (df["current-charge-degree"]=="Felony")
    df["current-charge-degree:Misdemeanor"] = (df["current-charge-degree"]=="Misdemeanor")
    
    return df

def drop_columns(df):
    drop_c = ['sex', 'age', 'juvenile-felonies', 'juvenile-misdemeanors',
       'juvenile-crimes', 'priors', 'current-charge-degree']
    df_drop = df.drop(drop_c, axis=1)
    
    return df_drop
```

```python
train = convert(train)
train = drop_columns(train)
train = train.astype("int")
print(train.shape)
train.head()
```

```python
train.to_csv("compas_train_preprocessed.csv")
```

```python
test = convert(test)
test = drop_columns(test)
test = test.astype("int")
print(test.shape)
test.head()
```

```python
test.to_csv("compas_test_preprocessed.csv")
```

## タームの生成

```python
with open("compas_train.out") as f:
    cond = f.read()
```

```python
terms = re.findall("{(.*)}", cond)
print(len(terms), terms)
```

```python
term_list = []
for i in terms:
    if "," in i:
        term = re.split(",", i)
        term_list.append({"feature": term, "value": np.ones(len(term), dtype=int)})
    else: term_list.append({"feature": [i], "value": [1]})
        
print(len(term_list), term_list)
```

```python
term_1 = []

for i in terms:
    if "," in i:
        term = re.split(",", i)
        df_temp = train[ (train[term[0]] == 1) & (train[term[1]] == 1) ]
    else:
        df_temp = train[train[i]==1]
        term = [i]
        
    label = df_temp["recidivate-within-two-years"].values

    if len(label) == 0: continue
    
    val = sum(label) / len(label)
    if val >= 0.5: 
        n_ones = np.ones(len(term), dtype=int)
        term_1.append({"feature": term, "value": n_ones})

print(len(term_1))
print(term_1)
```

```python
f = open('term_over1.pkl', 'wb')
pickle.dump(term_1, f)
f.close()
```
