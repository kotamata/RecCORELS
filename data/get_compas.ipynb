```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/propublica/compas-analysis/master/compas-scores-two-years.csv")
print(df.shape)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
columns = ["sex", "age", "juv_fel_count", "juv_misd_count", "juv_other_count", "priors_count", "c_charge_degree", "race", "two_year_recid"]
df_com = df[columns]
df_com.head()
```

```python
df_com.info()
```

```python
ages = []

for _, val in df_com["age"].iteritems():
    #print(row, val)
    if val <= 20:
        ages.append("18-20")
    elif val <= 22:
        #df_com["age"][row] = "21-22"
        ages.append("21-22")
    elif val <= 25:
        #df_com["age"][row] = "23-25"
        ages.append("23-25")
    elif val <= 45:
        #df_com["age"][row] = "26-45"
        ages.append("26-45")
    else:
        #df_com["age"][row] = ">45"
        ages.append(">45")
        
df_com.age = ages
df_com.head()       
```

```python
age_list = sorted(df_com.age.unique())
for i in age_list:
    print(i, len(df_com[df_com.age==i]))
```

```python
priors = []
p = df_com["priors_count"].values
#print(p)

for val in p:
    #print(row, val)
    if val == 0:
        priors.append("=0")
    elif val <= 1:
        #df_com["age"][row] = "21-22"
        priors.append("=1")
    elif val <= 3:
        #df_com["age"][row] = "23-25"
        priors.append("2-3")
    else:
        #df_com["age"][row] = ">45"
        priors.append(">3")
        
df_com["priors"] = priors
df_com.head()     
```

```python
pri_list = sorted(df_com.priors.unique())
for i in pri_list:
    print(i, len(df_com[df_com.priors==i]))
```

```python
j_fel = []
p = df_com["juv_fel_count"].values
#print(p)

for val in p:
    #print(row, val)
    if val == 0:
        j_fel.append("=0")
    else:
        #df_com["age"][row] = ">45"
        j_fel.append(">0")
        
df_com["juvenile-felonies"] = j_fel
df_com.head()  
```

```python
print("=0", len(df_com[df_com["juvenile-felonies"]=="=0"]))
print(">0", len(df_com[df_com["juvenile-felonies"]==">0"]))
```

```python
j_mis = []
p = df_com["juv_misd_count"].values
#print(p)

for val in p:
    #print(row, val)
    if val == 0:
        j_mis.append("=0")
    else:
        #df_com["age"][row] = ">45"
        j_mis.append(">0")
        
df_com["juvenile-misdemeanors"] = j_mis
df_com.head()
```

```python
print("=0", len(df_com[df_com["juvenile-misdemeanors"]=="=0"]))
print(">0", len(df_com[df_com["juvenile-misdemeanors"]==">0"]))
```

```python
j_cri = []
fel = df_com["juv_fel_count"].values
misd = df_com["juv_misd_count"].values
other = df_com["juv_other_count"].values
#print(p)

for i in range(len(fel)):
    #print(row, val)
    val = fel[i] + misd[i] + other[i]
    
    if val == 0:
        j_cri.append("=0")
    else:
        #df_com["age"][row] = ">45"
        j_cri.append(">0")
        
df_com["juvenile-crimes"] = j_cri
df_com.head()
```

```python
print("=0", len(df_com[df_com["juvenile-crimes"]=="=0"]))
print(">0", len(df_com[df_com["juvenile-crimes"]==">0"]))
```

```python
cl = []
c = df_com["c_charge_degree"].values
#print(p)

for val in c:
    #print(row, val)
    if val == "F":
        cl.append("Felony")
    else:
        #df_com["age"][row] = ">45"
        cl.append("Misdemeanor")
        
df_com["current-charge-degree"] = cl
df_com.head()
```

```python
column = ['sex', 'age', 'juvenile-felonies', 'juvenile-misdemeanors',
       'juvenile-crimes', 'priors', 'current-charge-degree', 'race', 
       'two_year_recid']
df_pre = df_com[column]
df_pre = df_pre.rename(columns={'two_year_recid': 'recidivate-within-two-years'})
df_pre.head()
```

```python
del_index = [7210, 7211, 7212, 7213]
print(del_index)
df_del = df_pre.drop(del_index)
df_del.tail()
```

```python
seed = sum([3, 15, 13, 16, 1, 19]) # c:3, o:15, m:13, p:16, a:1, s:19
np.random.seed(seed)
split_ind = np.split(np.random.permutation(len(df_del)), 10)
print(split_ind)
```

```python
df_test = df_del.iloc[split_ind[0]].reset_index()
df_test = df_test.drop("index", axis=1)

df_test
```

```python
df_train = df_del.iloc[split_ind[1]]
for i in np.arange(2, 10):
    df_add = df_del.iloc[split_ind[i]]
    df_train = pd.concat([df_train, df_add])

df_train = df_train.reset_index().drop("index", axis=1)
df_train
```

```python
df_train.to_csv("compas_train_race.csv")
df_test.to_csv("compas_test_race.csv")
```
